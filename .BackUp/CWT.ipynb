```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

# Generate a sample signal
t = np.linspace(0, 1, 1000)
signal = np.sin(50 * 2 * np.pi * t) + 0.5 * np.sin(80 * 2 * np.pi * t)

_Scales = 100
scales = np.arange(1, _Scales)

# Define the wavelet and scales
wavelet_lis = ['mexh','morl','gaus1','cmor1-1.0'] #
wavelet = wavelet_lis[1]  # You can choose other wavelet functions as well



for wavelet in reversed(wavelet_lis):
    # Perform Continuous Wavelet Transform
    print(wavelet)
    coefficients, frequencies = pywt.cwt(signal, scales, wavelet)

    # Plot the results
    plt.figure(figsize=(10, 6))
    plt.imshow(np.abs(coefficients), extent=[0, 1, 1, _Scales], aspect='auto', cmap='jet')
    plt.colorbar(label='Magnitude')
    plt.title('Continuous Wavelet Transform')
    plt.xlabel('Time')
    plt.ylabel('Scale')
    plt.show()
```
